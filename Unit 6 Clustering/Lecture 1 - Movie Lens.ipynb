```r
# After following the steps in the video, load the data into R
movies = read.table("movieLens.txt", header=FALSE, sep="|",quote="\"")

str(movies)
```

```r
# Add column names
colnames(movies) = c("ID", "Title", "ReleaseDate", "VideoReleaseDate", "IMDB", "Unknown", "Action", "Adventure", "Animation", "Childrens", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "FilmNoir", "Horror", "Musical", "Mystery", "Romance", "SciFi", "Thriller", "War", "Western")

str(movies)
```

```r
# Remove unnecessary variables
movies$ID = NULL
movies$ReleaseDate = NULL
movies$VideoReleaseDate = NULL
movies$IMDB = NULL

# Remove duplicates
movies = unique(movies)

# Take a look at our data again:
str(movies)
```

```r
# Compute distances
distances = dist(movies[2:20], method = "euclidean")
```

```r
str(distances)
```

```r
# Hierarchical clustering
clusterMovies = hclust(distances, method = "ward.D") 
```

```r
# Plot the dendrogram
plot(clusterMovies)
```

```r
# Assign points to clusters
clusterGroups = cutree(clusterMovies, k = 10)
```

```r
str(clusterGroups)
```

```r
tapply(movies$Action, clusterGroups, mean)
tapply(movies$Romance, clusterGroups, mean)
```

```r
# Assign points to clusters
clusterGroups2 = cutree(clusterMovies, k = 2)
```

```r
tapply(movies$Action, clusterGroups2, mean)
tapply(movies$Romance, clusterGroups2, mean)
```

```r
spl = split(movies[2:20], clusterGroups2)
lapply(spl, colMeans)

```
